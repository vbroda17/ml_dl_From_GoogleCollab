```python
# Vincent Broda, CS522 HW 1 problem 1-3-a

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# For class 1, no covid
avg1 = 95
sd1 = 5
prior1 = 0.8

# Class 2, yes covid
avg2 = 102
sd2 = 3
prior2 = 0.2

# Range of temp values for the graph
x = np.linspace(75, 115, 1000)

# we will do Bayes' Rule

# Get the conditional probability function, it is the likelyhood of it happening
# if we print this we get the gaussian curves
likely1 = norm.pdf(x, avg1, sd1)
likely2 = norm.pdf(x, avg2, sd2)

# we have the prior probability allready

# getting the normalization constants, or evidence in lectures slides
evidence = likely1 * prior1 + likely2 + prior2

# getting the posterior probabilities
post1 = likely1 * prior1 / evidence
post2 = likely2 * prior2 / evidence

# plotting the postiorior probability now
plt.plot(x, post1, label='Class 1 No Covid')
plt.plot(x, post2, label='Class 2 Yes Covid')
plt.xlabel('Temperature')
plt.title('Posterior Probability')

plt.legend()
plt.show()
```
