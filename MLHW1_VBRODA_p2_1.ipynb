```python
# Vincent Broda, CS522 HW 1 problem 2-1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
interval1 = [95, 100] # No covid
interval2 = [97, 103] # Yes covid

# rannge for x values shown (temperature in f)
x = np.linspace(92, 106, 1000)

# Getting pdfs of the unifrom distributions
pdf1 = uniform.pdf(x, loc=interval1[0], scale=interval1[1]-interval1[0])
pdf2 = uniform.pdf(x, loc=interval2[0], scale=interval2[1]-interval2[0])


# Ploting pdfs
plt.plot(x, pdf1, label='Class 1 No Covid')
plt.plot(x, pdf2, label='Class 2 Yes Covid')

# Adding other details to figure
plt.title('Uniform Probability Density Functions (PDFs)')
plt.xlabel("Temperature in Fahrenheit")
plt.ylabel("Probability Density")
#plt.grid(True)
plt.legend()
plt.show()

```
